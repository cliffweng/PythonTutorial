```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2*np.pi, 500)
plt.plot(x, np.sin(x**3))
plt.title('my graph')
```

```python
x = np.linspace(0, 50, 50)
y = x ** 2
plt.plot(x, y)
plt.title('my graph')
```
